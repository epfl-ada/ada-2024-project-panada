```python
from matplotlib.dates import DateFormatter, YearLocator
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
import pickle
```

```python
df_merged = pd.read_csv("merged_task2.csv")
```

```python
df_merged.head(10)
```

```python
df_merged["Engagement"] = (df_merged["num_comms"] + df_merged["like_count"] + df_merged["dislike_count"])/(df_merged["view_count"])
```

```python
df_merged.head(10)
```

```python
df_new = df_merged.drop(columns=['num_comms', 'like_count', 'video_id', 'dislike_count', 'like_count', 'upload_date', 'view_count', 'display_id'])
```

```python
df_new.head(10)
```

```python
df_dummies_merged = pd.get_dummies(df_new, columns=['category'], drop_first=True)
```

```python
df_dummies_merged = df_dummies_merged.sample(frac=1, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

```python
df_dummies_merged.dropna(inplace = True)
```

```python
df_LR = df_dummies_merged[:10000]
```

```python
# Variables explicatives (features) et variable cible (target)
X = df_LR.drop(columns=["Engagement", "duration"])  # Toutes les colonnes sauf Engagement
y = df_LR["Engagement"]

# Ajout d'une constante (nécessaire pour l'interception dans statsmodels)
X = sm.add_constant(X)

# Modèle de régression linéaire
model = sm.OLS(y, X).fit()

# Résultats
print(model.summary())
```
